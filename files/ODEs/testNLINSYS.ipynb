```python
import numpy as np
from quasi_nwt import newton_solver, quasi_newton_solver


def F_3D(x):
    return np.array([np.cos(x[0]), np.sin(x[1]), x[2]**2 - 1])


FF=lambda x: np.array([np.cos(x[0]), np.sin(x[1]), x[2]**2 - 1])

def J_3D(x):
    return np.array([
        [-np.sin(x[0]), 0, 0],
        [0, np.cos(x[1]), 0],
        [0, 0, 2*x[2]]
    ])

JJ=lambda x: np.array([
        [-np.sin(x[0]), 0, 0],
        [0, np.cos(x[1]), 0],
        [0, 0, 2*x[2]]
    ])


solution_newton, iterations_newton = newton_solver([0.5, -1, 0.3], FF, J_3D, 1e-10, 100)
print(f"Newton Solution: {solution_newton}, Iterations: {iterations_newton}")
print(f"F at solution: {F_3D(solution_newton)}")
solution_quasi, iterations_quasi = quasi_newton_solver([0.5, -1, 0.3], F_3D, 1e-10, 1000)
print(f"Quasi-Newton Solution: {solution_quasi}, Iterations: {iterations_quasi}")
print(f"F at solution: {F_3D(solution_quasi)}")
```
